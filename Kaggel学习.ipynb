```python
#列表推导式
squares=[i**2 for i in range(10)]
print(squares)
```

```python
# 带条件的列表推导式：
evens=[i for i in range(10) if i%2==0]
print(evens)
labels=['pass' if score>=60 else 'fail' for score in [85,45,92,58]]
print(labels)
```

常用内置函数

```python
numbers=[3,1,4,1,5,9,2,6]

sum(numbers)        #求和
max(numbers)        #最大值
min(numbers)        #最小值
len(numbers)        #长度
sorted(numbers)     #排序

#enumerate - 带索引遍历
for idx,val in enumerate(numbers):
    print(f"第{idx}个元素是{val}")

# zip - 并行遍历多个列表
names=['Alice','Bob','Charlie']
scores=[85,92,78]
for name,scores in zip(names,scores):
    print(f"{name}:{scores}分")

# map - 对每个元素应用函数
squared=list(map(lambda x:x**2,[1,2,3]))
print(squared)
# filter - 过滤元素
big=list(filter(lambda x:x>5,[1,3,6,8,2]))
print(big)
```

```python
#1.函数参数进阶
#(1)默认参数——不传就用默认值
def train_model(epochs=10,learning_rate=0.001,batch_size=32):
    print(f"训练{epochs}轮，学习率={learning_rate}，批大小={batch_size}")
train_model()                                #使用默认参数,{10，0.001，32}
train_model(epochs=20)                       #覆盖默认参数,{20，0.001，32}
train_model(learning_rate=0.0005,epochs=30)  #覆盖部分,{30，0.0005，32}
#(2)args——接收任意数量的位置参数
def calculate_sum(*args):
    """接收任意多个数字，返回它们的和"""
    print(f"接收到了{len(args)}个参数：{args}")
    return sum(args)
print(calculate_sum(1,2,3))
print(calculate_sum(10,20,30,40))
#(3)kwargs_接收任意数量的关键字参数
def print_model_confiig(**kwargs):
    """接收任意多个命名参数，打印模型配置"""
    print("模型配置：")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
print_model_confiig(layers=3,units=128,activation="relu",dropout=0.2)
```

```python
# 文本读写操作
##重要：始终使用 with open(...) as f: 语法，它会自动关闭文件，避免资源泄漏！
#(1)读写文本文件
##写入文件（'w'模式会覆盖，'a'模式会追加）
with open('output.txt','w',encoding='utf-8')as f:
    f.write("第一行数据\n")
    f.write("第二行数据\n")
    f.writelines(["第三行数据\n","第四行数据\n"])
##读取文件
with open('output.txt','r',encoding='utf-8')as f:
    content=f.read()    #读取全部内容（字符串）
    print(content)
with open('output.txt','r',encoding='utf-8')as f:
    lines=f.readlines() #读取为列表，每行一个元素
    print(lines)        #['第一行数据\n', '第二行数据\n',...]
#逐行读取（推荐用于大文件）
with open('output.txt','r',encoding='utf-8')as f:
    for line in f:
        print(line.strip())  #strip()用于去除换行符
```

```python
# 任务1：写一个函数，接收任意数量的数字，返回平均值

# 任务：实现 average 函数
# 要求：
#   - 使用 *args 接收参数
#   - 返回所有数字的平均值（浮点数）
#   - 如果没有传入参数，返回 0

def average(*args):
  return sum(args)/len(args) if args else 0

# 测试用例
print(average(1, 2, 3, 4, 5))       # 期望输出: 3.0
print(average(10, 20, 30))          # 期望输出: 20.0
print(average(85, 92, 78, 90, 88))  # 期望输出: 86.6
print(average())                     # 期望输出: 0
```

```python
#任务 2：读取 CSV 文件，计算指定列的平均值

#先运行以下代码创建一个测试用的 CSV 文件
import csv

# 创建测试数据
with open('test_scores.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['姓名', '数学', '英语', '科学'])
    writer.writerow(['Alice', '85', '92', '78'])
    writer.writerow(['Bob', '90', '88', '95'])
    writer.writerow(['Charlie', '76', '85', '82'])
    writer.writerow(['Diana', '92', '79', '91'])
    writer.writerow(['Eve', '88', '94', '87'])

import pandas as pd

# 任务：实现 column_average 函数
# 要求：
#   - 读取指定 CSV 文件
#   - 计算指定列的平均值
#   - 返回浮点数结果

def column_average(filename, column_name):
  df=pd.read_csv(filename)
  return df[column_name].mean()

# 测试用例
math_avg = column_average('test_scores.csv', '数学')
print(f"数学平均分: {math_avg}")    # 期望输出: 86.2

english_avg = column_average('test_scores.csv', '英语')
print(f"英语平均分: {english_avg}")  # 期望输出: 87.6

science_avg = column_average('test_scores.csv', '科学')
print(f"科学平均分: {science_avg}")  # 期望输出: 86.6
```

```python
#任务 3：将学生信息写入 CSV 文件
import csv

# 任务：实现 save_students 函数
# 要求：
#   - 接收学生列表（每个学生是字典）
#   - 将数据保存为 CSV 文件
#   - 第一行是表头（从字典的 key 获取）
#   - 使用 csv.DictWriter

students = [
    {'name': 'Alice', 'score': 85, 'grade': 'A'},
    {'name': 'Bob', 'score': 92, 'grade': 'A'},
    {'name': 'Charlie', 'score': 78, 'grade': 'B'},
    {'name': 'Diana', 'score': 96, 'grade': 'A+'},
]

def save_students(students, filename):
  if not students:
    return
  # 修正 1: 使用 .keys() 而不是 .key()
  fieldnames = list(students[0].keys())
  with open(filename, 'w', newline='', encoding='utf-8') as f:
    # 修正 2: 确保变量名 fieldnames 一致
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(students)


# 测试
save_students(students, 'my_students.csv')

# 验证：读取并打印
import pandas as pd
df = pd.read_csv('my_students.csv')
print(df)
# 期望输出：
#      name  score grade
# 0    Alice   85   A
# 1    Bob    92   A
# 2    Charlie  78   B
# 3    Diana   96   A+
```
